```
import requests as r
import csv
import pandas as pl
import matplotlib.pyplot as plt


url = 'https://api.covid19api.com/dayone/country/'
pais = (input('digite o pais\n'))
resp = r.get(url+pais)
raw_data = resp.json()
raw_data[0]
final_data = []
for obs in raw_data:
  final_data.append([obs['Country'], obs['Confirmed'], obs['Deaths'], obs['Recovered'], obs['Active'], obs['Date']])    
final_data.insert(0,['Pais', 'confirmados', 'obitos', 'recuperados', 'ativos', 'data'])
final_data
with open ('dados_mundiaiscovid19.csv', 'w') as file:
  writer = csv.writer(file)
  writer.writerows(final_data)
dados = pl.read_csv('/content/dados_mundiaiscovid19.csv')
plt.figure(figsize=(8, 6))
plt.plot(dados['confirmados'], dados['obitos'])
plt.xlabel('obitos')
plt.ylabel('confirmados')
plt.title('dados covid19 por pais')
plt.savefig('dadosmundo.png')


```
